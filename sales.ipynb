```python
sales['Date'] = pd.to_datetime(sales['Date'])
sales

sales['Year']=sales['Date'].dt.year
sales['Month']=sales['Date'].dt.month
sales['Day']=sales['Date'].dt.day

sales
```

```python
sales=sales.drop(['Date','Pages_Viewed'],axis=1)
sales
```

```python
sales['Year'].unique()
```

```python
sales.isnull().sum()
sales.isnull()
```

```python
sales.drop_duplicates(inplace=True)
sales
```

```python

sales['Unit_Price'] = pd.to_numeric(sales['Unit_Price'], errors='coerce')

sales
```

```python
sales['City'] = sales['City'].str.strip().str.title()
sales['City']
```

```python
sales = sales[sales['Age'].between(15, 90)]
sales['Age'].unique()
sales
```

```python
sales['Total_Match'] = abs((sales['Unit_Price']  * sales['Quantity'] )
-sales['Discount_Amount']- sales['Total_Amount']) < 0.01

```

```python
sales.loc[sales['Total_Match']==False,'Total_Amount']    #=sales.loc[sales['Total_Match']==False,'Unit_Price']
print("\/ All Total are corrected succesfully")
sales['Total_Match'].value_counts()
sales
```

```python
sales['Cost']=sales['Unit_Price']*0.7
sales['Profit']=(sales['Total_Amount']-(sales['Cost']*sales['Quantity']))
sales
```

```python
sales.to_csv("Cleanedsales.csv", index=False)

print("Cleaned data saved successfully as 'Cleanedsales.csv'")
sales
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
# sales = pd.read_csv(r"C:\Users\gsudh\OneDrive\Documents\Excel\ecommerce_customer_behavior_dataset.csv")

# Create figure and subplots
plt.figure(figsize=(14, 10))

# 1️⃣ Bar chart – Number of sales by city
plt.subplot(2, 2, 1)
sales['City'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Cities by Number of Sales')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=45)

# 2️⃣ Pie chart – Payment method distribution
plt.subplot(2, 2, 2)
sales['Payment_Method'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Payment Method Distribution')
plt.ylabel('')

# 3️⃣ Histogram – Distribution of Total Amount
plt.subplot(2, 2, 3)
plt.hist(sales['Total_Amount'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Total Amount')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')

# 4️⃣ Scatter Plot – Age vs Total Amount
plt.subplot(2, 2, 4)
avg_duration = sales.groupby('Device_Type')['Session_Duration_Minutes'].mean()
avg_duration.plot(kind='line', marker='o', color='purple')
plt.title('Average Session Duration by Device Type')
plt.xlabel('Device Type')
plt.ylabel('Average Duration (minutes)')


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


pivot_data = sales.pivot_table(
    index='City',
    columns='Product_Category',
    values='Profit',
    aggfunc='sum'
)

# Create heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_data, annot=True, fmt=".0f", cmap='YlGnBu')

plt.title('Sales Profit Heatmap by City and Product Category')
plt.xlabel('Product Category')
plt.ylabel('City')
plt.show()
```

```python
month_profit = sales.groupby('Month')['Profit'].sum().cumsum()

plt.fill_between(month_profit.index, month_profit.values, color='orange', alpha=0.5)
plt.title('Cumulative Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Cumulative Profit')
plt.show()
```

```python
city_sales = sales.groupby('City')['Profit'].sum()
plt.pie(city_sales, labels=city_sales.index, autopct='%1.1f%%', startangle=90, wedgeprops={'width':0.3})
plt.title('Sales Share by City (Donut Chart)')
plt.show()
```
